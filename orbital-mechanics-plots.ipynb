```python
# Importing numpy and matplotlib
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
file_name = 'circular_motion.dat' 
all_data = np.loadtxt(file_name,skiprows=1) 
all_data = all_data.transpose() 
time = all_data[0] 
x1 = all_data[1] 
y1 = all_data[2] 
x2 = all_data[3] 
y2 = all_data[4] 
energy = all_data[5] 
```

```python
plt.plot(x1, y1, label = "planet 1 with mass 1, position 1, 0 and velocity 0, 1")
plt.plot(x2, y2, label = "planet 2 with mass 1, position -1, 0 and velocity 0, -1")
plt.xlabel('x position in meters with primary mass of mass 1 at origin')
plt.ylabel('y position in meters') 
plt.legend()
plt.show()
```

```python
plt.plot(time, energy, label = "energy") 
plt.xlabel('time in seconds') 
plt.ylabel('energy') 
plt.legend()
plt.show()
```
